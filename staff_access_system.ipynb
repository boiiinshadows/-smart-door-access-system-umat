# Case Study: Smart Door Access System for UMaT

**Course:** EL 162/234
**Task:** Build a `Staff` class for the smart door access system used by UMaT admin staff.

- Public attribute: `s_name`
- Private attribute: `__access_code`
- Methods to:
  - View the access code (authorized users only)
  - Update the access code after validating its length
  - Display staff information
- Uses `@property` decorator instead of traditional getter/setter methods


```python
class Staff:

    def __init__(self, s_name, access_code):
        self.s_name = s_name
        self.__access_code = access_code

    @property
    def access_code(self):
        return self.__access_code

    @access_code.setter
    def access_code(self, new_code):
        if not isinstance(new_code, str):
            raise TypeError("Access code must be a string.")
        if len(new_code) < 4:
            raise ValueError("Access code must be at least 4 characters long.")
        self.__access_code = new_code
        print(f"Access code for {self.s_name} updated successfully.")

    def display_info(self):
        masked = "*" * len(self.__access_code)
        print(f"Staff Name: {self.s_name}")
        print(f"Access Code: {masked}")
```

## Demo / Test

```python
staff1 = Staff("Edwin Adom", "AD1234")
staff1.display_info()
```

```python
print("Current code:", staff1.access_code)
```

```python
staff1.access_code = "NEW5678"
staff1.display_info()
```

```python
try:
    staff1.access_code = "abc"
except ValueError as e:
    print("Error:", e)
```

## Notes

- `s_name` is a plain public attribute.
- `__access_code` is name-mangled by Python to `_Staff__access_code`, making it private.
- The `@property` decorator exposes `access_code` as a controlled attribute: reading it calls the getter, assigning to it calls the setter.
- The setter validates type and minimum length before allowing an update.
- `display_info()` never prints the raw code, only a masked version.
